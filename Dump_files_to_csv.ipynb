```python
# Rxtract text from wiki dump files
def Extract_text(file_path):
  title=""
  para_content=""
  with open(file_path) as fp:
    soup=BeautifulSoup(fp, "html.parser")
    # We have to use if statement because some files doesn;t have title
    if soup.title:
      title=soup.title.get_text()
    paragraph = soup.find_all('p')
    para_content = reduce((lambda x,y:x+" "+y.text), paragraph,"")
  return {'title': title, 'para_content':para_content}
```

```python
#os.fsencode return the file system representation of the path.
positive_data = os.fsencode('/content/drive/MyDrive/The Problem: Classify Wikipedia Disease Articles/training/positive')
negative_data = os.fsencode('/content/drive/MyDrive/The Problem: Classify Wikipedia Disease Articles/training/negative')
```

```python
#f1 = os.listdir(positive_data)[0]
#print(os.path.join(positive_data, f1))
#Extract_text(os.path.join(positive_data, f1))
```

```python
def Extract_data(data, data_type):
  data_title=[]
  data_content=[]
  y=[]

  for file in os.listdir(data):
    file_path = os.path.join(data, file)
    extracted_text = Extract_text(file_path)
    data_title.append(extracted_text['title'])
    data_content.append(extracted_text['para_content'])
    y.append(data_type)
  
  data_frame={'data_title':data_title, 'data_content':data_content, 'y':y}
  df=pd.DataFrame(data_frame)
  return df
```
